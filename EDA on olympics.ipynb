```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns




df=pd.read_csv("medal.csv",na_values=['Missing',' '])
print(df.head())
df.describe()
```

```python
def find(year,sport,Medal="Gold"):
    data_find=df[(df['Year']==year) & (df['Sport']==sport) & (df['Medal']==Medal)]
    return data_find
find(2000,"Wrestling")
```

```python
df['Gender'].value_counts().plot(kind='barh', figsize=(8, 6))
```

```python
df.groupby(['Year','Gender']).size().unstack().plot(figsize=(15,10),kind='bar',stacked=True)
plt.show()
```

```python
print(df.groupby(['City','Medal']).size().reset_index(name='Total_medals'))
ax=df.groupby(['City','Medal']).size().unstack().plot(figsize=(17,10),kind='bar',stacked=True,title='Cities with all medals',color=('chocolate','gold','silver'))

ax.set_ylabel("Total Medals",fontsize=18)
ax.set_xlabel("Cities",fontsize=18)
plt.show()
```

```python

ax=df.groupby(['City','Gender']).size().unstack().plot(figsize=(15,10),kind='barh',stacked=True,title='Men/Women ratio')
ax.set_xlabel("Total participants",fontsize=18)
ax.set_ylabel("Cities",fontsize=18)
plt.show()
```
